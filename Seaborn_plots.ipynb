```python
import seaborn as sns
```

```python
#loadking data set
t = sns.load_dataset('titanic')
i = sns.load_dataset('iris')
p = sns.load_dataset('penguins')
tip = sns.load_dataset('tips')
f = sns.load_dataset('flights')
fmri = sns.load_dataset("fmri")
```

#1.Scatter plot

  * USE : visualizing theh relationship between two continuous variables
  * USE CASE  : analyzing the correlations and patterns between the data points
  * sns.scatterplot(x=data['x'], y=data['y'])

```python
#scatter plot
sns.scatterplot(data = tip,x = 'total_bill' , y = 'tip')
```

```python
sns.scatterplot(data = tip, x = 'total_bill' , y = 'tip',style = 'time',hue = 'sex')
```

```python
sns.scatterplot(data = tip, x = 'total_bill' , y = 'tip',style = 'time',hue = 'day')
```

#2.lineplot
* Use: Showing the trend or progression of a continuous variable over a specific interval.

* Use Case: Displaying time-series data or continuous data trends.
* Example: sns.lineplot(x=data['time'], y=data['value'])

```python
sns.lineplot(data = tip, x = 'day',y = 'tip')
```

```python
sns.lineplot(data = f,x = 'year',y = 'passengers',hue = 'month')
```

```python
sns.lineplot(data = fmri,x = 'timepoint',y = 'signal',hue = 'event',style='event',markers=True,dashes=False)
```

#3.BARPLOT
* Use: Comparing discrete categories with their respective values.
* Use Case: Comparing sales by product categories or survey responses.
* Example: sns.barplot(x=data['categories'], y=data['values'])


```python
sns.barplot(data = p,x = 'island',y = 'body_mass_g',hue = 'sex')
```

```python
sns.barplot(data = t ,x = 'class',y = 'age', hue = 'sex')
```

#4.CountPlot
* Use: Counting the occurrences of categorical variables.
* Use Case: Displaying the distribution of categorical data.
* Example: sns.countplot(x=data['categories'])

```python
sns.countplot(data = t,x = 'survived',hue = 'sex')
```

```python
sns.countplot(data =t,x ='class',hue='sex')
```

```python
sns.countplot(data=t, y="deck", hue="alive")
```

```python
sns.countplot(data = p,x = 'sex',hue='island')
```

#5. HistPlot
* Use: Visualizing the distribution of continuous variables.
* Use Case: Analyzing the frequency distribution of data.
* Example: sns.histplot(data['values'], bins=10)

```python
sns.histplot(data = t,x ='fare',bins=[10,50,100,150,200])
```

```python
sns.histplot(data=p, x="flipper_length_mm")
```

```python
sns.histplot(data=p, x="flipper_length_mm", bins=30)
```

```python
sns.histplot(data=i,palette='Wistia')
```

#6. BOX Plot
* Use: Displaying the summary statistics of a continuous variable's distribution.
* Use Case: Identifying outliers and comparing distributions.
* Example: sns.boxplot(x=data['group'], y=data['values'])

```python
sns.boxplot(data =t,x = t['age'])
```

```python
sns.boxplot(data=t,x = 'age',y = 'class')
```

```python
sns.boxplot(data=t,x = 'age',y = 'class',hue = 'alive')
```

#7. Voilin Plot
* Use: Combines a box plot with a kernel density estimation for better distribution insight.
* Use Case: Visualizing distribution and summary statistics together.
* Example: sns.violinplot(x=data['group'], y=data['values'])

```python
sns.violinplot(data =t,x ='class',y = 'age')
```

```python
sns.violinplot(data = i, x='species',y ='sepal_length',inner="stick")
```

```python
sns.violinplot(data=t, x="deck", y="age", hue="alive", split=True)
```

#8. Pairplot
* Use: Creating scatter plots between pairs of variables in a dataset.
* Use Case: Exploring relationships and distributions across multiple variables.
* Example: sns.pairplot(data)

```python
sns.pairplot(data=i)
```

```python
sns.pairplot(data = f)
```

```python
sns.pairplot(data=p, hue="species",palette='Reds')
```

#9. jointplot
* Use: Combining two different visualizations for bivariate data.
* Use Case: Analyzing the relationship between two variables with scatter plots and marginal distributions.
* Example: sns.jointplot(x=data['x'], y=data['y'], kind='scatter')

```python
sns.jointplot(data=p, x="bill_length_mm", y="bill_depth_mm")
```

```python
sns.jointplot(data=p, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=i,kind='hex')
```

#10. KDE PLOT
* Use: Displaying the distribution of a single continuous variable as a smooth curve.
* Use Case: Visualizing the probability density function of data.
* Example: sns.kdeplot(data['values'])

```python
sns.kdeplot(data=i)
```

```python
sns.kdeplot(data=tip, x="total_bill", hue="time")
```

```python
sns.kdeplot(
   data=tip, x="total_bill", hue="size",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0,
)
```

#11. Strip Plot
* Use: Displaying individual data points as dots along a categorical axis.
* Use Case: Visualizing the distribution of data points within categories.
* Example: sns.stripplot(x=data['categories'], y=data['values'])

```python
sns.stripplot(x='species',y='body_mass_g',data=p)
```

```python
sns.stripplot(x='species',y='body_mass_g',data=p,hue='sex')
```

#12. DISPLOT
*This function provides access to several approaches for visualizing the univariate or bivariate distribution of data, including subsets of data defined by semantic mapping and faceting across multiple subplots.
* seaborn.displot(data=None, *, x=None, y=None)

```python
sns.distplot(a=t.age)
```

```python
sns.displot(data=t,x='age')
```

#13. HEATMAP
* Use: Displaying a matrix dataset as a grid of colored squares.
* Use Case: Visualizing correlations, pairwise relationships, and matrices.
* Example: sns.heatmap(data.corr(), annot=True)

```python
glue = sns.load_dataset("glue").pivot("Model", "Task", "Score")
sns.heatmap(glue)
```

```python
sns.heatmap(glue, annot=True)
```
